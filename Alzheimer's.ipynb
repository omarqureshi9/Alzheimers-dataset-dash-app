```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

# Load the data
df = data

# Create a list of unique countries from the DataFrame
countries = df['Entity'].unique()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Death Rates by Country"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in countries],
        value=countries[0]  # Default value
    ),
    dcc.Graph(id='death-rates-graph')
])

# Define the callback to update the graph based on the selected country
@app.callback(
    Output('death-rates-graph', 'figure'),
    Input('country-dropdown', 'value')
)
def update_graph(selected_country):
    filtered_df = df[df['Entity'] == selected_country]
    fig = {
        'data': [{
            'x': filtered_df['Year'],
            'y': filtered_df['Deaths'],
            'type': 'line'
        }],
        'layout': {
            'title': f'Death Rates for {selected_country}',
            'xaxis': {'title': 'Year'},
            'yaxis': {'title': 'Death Rate'}
        }
    }
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import dash_table
import pandas as pd

# Load the data
df = data

# Create a list of unique countries from the DataFrame
countries = df['Entity'].unique()

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("Death Rates by Country"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in countries],
        value=countries[0]  # Default value
    ),
    dcc.Graph(id='death-rates-graph'),
    dash_table.DataTable(
        id='death-rates-table',
        columns=[
            {'name': 'Year', 'id': 'Year'},
            {'name': 'Deaths', 'id': 'Deaths'}
        ],
        data=[],  # This will be filled by the callback
        style_table={'overflowX': 'auto'},
        style_cell={'textAlign': 'left'},
    )
])

# Define the callback to update the graph and table based on the selected country
@app.callback(
    [Output('death-rates-graph', 'figure'),
     Output('death-rates-table', 'data')],
    [Input('country-dropdown', 'value')]
)
def update_output(selected_country):
    filtered_df = df[df['Entity'] == selected_country]
    fig = {
        'data': [{
            'x': filtered_df['Year'],
            'y': filtered_df['Deaths'],
            'type': 'line'
        }],
        'layout': {
            'title': f'Death Rates for {selected_country}',
            'xaxis': {'title': 'Year'},
            'yaxis': {'title': 'Death Rate'}
        }
    }
    table_data = filtered_df[['Year', 'Deaths']].to_dict('records')
    return fig, table_data

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
